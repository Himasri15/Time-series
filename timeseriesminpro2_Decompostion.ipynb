In this min project we are going to know about Decomposition in Time Seires.
In time series we have ETS(Error, Trend, Seasonality) models. Here we done on ETS Decompostion which is specific model in ETS.
Stats models provides a seasonal decomposition tool we can use to separate the different components.


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('/content/drive/MyDrive/Datascience/Time series 14-12-2020/airline_passengers.csv',index_col='Month',parse_dates=True)
```

```
df.info()
```

```
df = df.dropna()
```

```
df.head()
```

```
import matplotlib.pyplot as plt
```

```
df.plot()
```

In the graph we can observe the linear is little high so we are going for exponential(multiplicative, which means its is nonlinear).A couple of key thing when performing ETS Decomposition.They are two imp models in ETS a) Additive - we apply additive when models seems to linear b) multiplicative - we apply multiplicative model is more appropriate which we are increasing or decreasing at a non linear(exponential)

```
from statsmodels.tsa.seasonal import seasonal_decompose
```

```
result = seasonal_decompose(df['Thousands of Passengers'],model='multiplicative')
```
